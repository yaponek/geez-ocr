```python
# All Imports
import os
import glob
import PIL
import cv2
import numpy as np
import matplotlib
import matplotlib.cm as cm
from matplotlib import pyplot as plt
import random


#  All Utility Methods
def view(img):
    plt.imshow(img, cmap='Greys_r')
    plt.show()

def save(ImgName, Img):
    matplotlib.image.imsave(ImgName, Img, cmap='Greys_r')

def rand_col_box(image, initial, final, color_interval, tickness):
    # # this method draw and 
    # get a copy of the image
    tmp = np.copy(image)
    # randomly Generate a color set
    red = random.randrange(0, 255, color_interval)
    green = random.randrange(0, 255, color_interval)
    blue = random.randrange(0, 255, color_interval)
    # draw box over it then return
    x,y = initial
    w,h = final
    return cv2.rectangle(tmp, (x,y), ((x+w),(y+h)), (red, green, blue), tickness)



def grayImg(image):
    # this method return gray version of the given image
    return cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

def blackWhiteImg(grayscaleImage):
    return cv2.threshold(grayscaleImage, 128, 255, cv2.THRESH_BINARY)[1]

def get_file_name(arg):
    # # this method get all the file names of evry file in a given path or folder
    path = arg[0]
    files = glob.glob(path+'/'+ '*.*')
    size = len(files);
    if(arg[2]<size):
        if(arg[1]=='next'):
            file_data = files[arg[2]]
        elif(arg[1]=='priv'):
            file_data = files[arg[2]]
        else:
            return 'Error: Unknown command '+arg[1]+", use [next] or [priv] only",size
            
    else:
        return 'Error: out of range, only '+str(size)+" available",size
    # return array with 1 file name @ index 0  & total number of files @ index 1 
    return file_data,size

def bounder(img, initial, final):
    # # This method make a nose removal and collor-full rectangular border
    # make it gray
    gray= grayImg(img)
    # geting the threshold
    ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
    # noise removal
    kernel = np.ones((3,3),np.uint8)
    opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2)
    # sure background area
    sure_bg = cv2.dilate(opening,kernel,iterations=3)
    # contours_process(sure_bg, img)

    return rand_col_box(img, initial, final, 30, 1)

def fidel(bet):
    # this method sort and prepare the geez table for standard use from unicode
    deqala1 = ['ቈ','ኈ','ኰ','ዀ','ጐ']
    deqala2 = ['ሏ','ሗ','ሟ','ሧ','ሯ','ሷ','ሿ','ቊ','ቧ','ቯ',
               'ቷ','ቿ','ኊ','ኗ','ኟ','ኧ','ኲ','ዂ','ዟ','ዧ',
               'ዷ','ጇ','ጒ','ጧ','ጯ','ጷ','ጿ','ፏ','ፗ']
    deqala3 =['ቋ','ኋ','ኳ','ዃ','ጓ']
    deqala4 = ['ቌ','ኌ','ኴ','ዄ','ጔ']
    deqala5 = ['ቍ','ኍ','ኵ','ዅ','ጕ']
    
    i=0;fidel = 4608;fideloch = []
    if bet==1: fidel = 4608
    elif bet==2: fidel = 4609
    elif bet==3: fidel = 4610
    elif bet==4: fidel = 4611
    elif bet==5: fidel = 4612
    elif bet==6: fidel = 4613
    elif bet==7: fidel = 4614
    elif bet==8: 
        for word in deqala1: fideloch.append(word)
    elif bet==9: 
        for word in deqala2: fideloch.append(word)
    elif bet==10: 
        for word in deqala3: fideloch.append(word)
    elif bet==11: 
        for word in deqala4: fideloch.append(word)
    elif bet==12: 
        for word in deqala5: fideloch.append(word)
            
    if bet < 8:
        while i<34:
            fideloch.append(chr(fidel))
            if i==8: fidel = fidel+32
            elif i==13 or i==17 or i==24: fidel = fidel+16
            elif i==18: fidel = fidel+16
            elif i==26: fidel = fidel+24
            else: fidel = fidel+8
            i=i+1
    return fideloch




#------------------------------------------------------------------------------------------





# Declare a list variable witl all the geez laters in ot according to our form standard
fidels = []
for index in range(1,13):
    fidels.append(fidel(index))

# fixed standard height and Width value of form pre-processor
h_w = [[25,890], [25,890], [25,890], [25,890], [25,890], [25,890], [25,890],# from geez to sabe
       
       [25,25], [25,25], [25,25], [25,25], [25,25],  # 1st column of diqala qalat
       [25,480], [25,50], [25,175], [25,50], # 2nd column of diqala qalat
       [25,25], [25,25], [25,25], [25,25], [25,25],  # 3rd column of diqala qalat
       [25,25], [25,25], [25,25], [25,25], [25,25],  # 4th column of diqala qalat
       [25,25], [25,25], [25,25], [25,25], [25,25]]  # 5th column of diqala qalat

#len(x_y)=31
# Adjustable y coordinate value of each later from selected Column of the form pre-processor
           #  ሀ1 ለ2  ሐ3  መ4   ሰ5  ረ6   ሰ7   ሸ8   ቀ9   በ10
initial = [0, 0, 25, 50, 75, 100, 125, 150, 175, 200, 225, 
           #ቨ11 ተ12  ቸ13  ኀ14  ነ15  ኘ16  አ17  ከ18  ኸ19  ወ20 
           250, 275, 300, 325, 350, 375, 400, 425, 450, 475,
           #ዐ21 ዘ22  ዠ23  የ24  ደ25  ጀ26  ገ27  ጠ28  ጨ29 ጰ30  
           500, 525, 550, 575, 600, 625, 650, 675, 700, 725,
           #ጸ31 ፀ32  ፈ33  ፐ34
           750, 775, 800, 825]  # 5th column of diqala qalat

           #  # ሏ30 ሗ31 ሟ32 ሧ33 ሯ34  ሷ35  ሿ36  ቊ37  ቧ38 ቯ39 ቷ40  ቿ41 ኊ42  ኗ43  ኟ44
initial_2 = [0, 25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325, 350, 375,
            #ኧ45 ኲ46 ዂ47  ዟ48  ዧ49 ዷ50  ጇ51 ጒ52  ጧ53 ጯ54  ጷ55 ጿ56  ፏ57  ፗ58  
            400, 425, 450, 525, 550, 600, 625, 650, 675, 700, 725, 750, 800, 825]

#------------------------------------------------------------------------------------------


# Adjustable x-y coordinate value of form pre-processor Column
x_y = [[2,0], [62,0], [120,0], [177,0], [237,0], [292,0], [351,0],# from geez to sabe [->,^]
       
       [408,214], [406,346], [408,451], [407,478], [409,686], # 1st column of diqala qalat
       [466,31], [475,558], [471,640], [471,840],  # 2nd column of diqala qalat
       [528,216], [522,346], [529,452], [529,478], [525,687],  # 3rd column of diqala qalat
       [584,217], [584,348], [586,450], [586,476], [586,688],  # 4th column of diqala qalat
       [649,216], [644,345], [646,450], [646,476], [648,687]]  # 5th column of diqala qalat

# initiate segmentation row pading valuse

index = 1; dynamic_pad=0; constant_pad = 2;

while index < len(initial):
    if index==2:
        dynamic_pad=0
    elif index==6:
        dynamic_pad=5
    initial[index] = initial[index]+dynamic_pad+constant_pad
    index=index+1

# initiate segmentation row pading valuse of initial_2

index = 1; dynamic_pad=0; constant_pad = 2;

while index < len(initial_2):
    if index==6:
        dynamic_pad=5
    elif index==19:
        dynamic_pad=10
    initial_2[index] = initial_2[index]+dynamic_pad+constant_pad
    index=index+1
    
    
#------------------------------------------------------------------------------------------

# # Fetch file
form_number = 132
path ='datasetl\\'
file_name ='form '+str(form_number)+'.png'
# Load an color image in grayscale
imx = cv2.imread(path+file_name)
    
i=0; x2=1

while i < len(x_y):
    
    # a critical fix to match the icroped image content with exact diqala character unicode representation
    fidel_index_fix = 7
    if i in range(7,12):
        fidel_index_fix=7
    elif i in range(12,16):
        fidel_index_fix=8
    elif i in range(16,21):
        fidel_index_fix=9
    elif i in range(21,26):
        fidel_index_fix=10
    elif i in range(26,31):
        fidel_index_fix=11
    
    # the main itration of bounding and/or croping
    if h_w[i][1]==890:
        x=0
        while x < 34:
            imx = bounder(imx, [x_y[i][0],initial[x+1]+x], [25,25])
            x=x+1
    elif h_w[i][1]==480:
        x=0
        while x < 18:
            imx = bounder(imx, [x_y[i][0],initial_2[x+1]+x], [25,25])
            x=x+1
    elif  h_w[i][1]==175:
        x=20
        while x < 29:
            imx = bounder(imx, [x_y[i][0],initial_2[x+1]+x], [25,25])
            x=x+1
    elif h_w[i][1]==50 and x_y[i][1]==558:
        x=18
        while x < 20:
            imx = bounder(imx, [x_y[i][0],initial_2[x+1]+x], [25,25])
            x=x+1
    elif h_w[i][1]==25:
        if x2 < 6:
            imx = bounder(imx,[x_y[i][0],x_y[i][1]], [25,25])
            x2=x2+1
        else:
            x2=1
            imx = bounder(imx,[x_y[i][0],x_y[i][1]], [25,25])
            x2=x2+1
            
    i=i+1
imx = cv2.resize(imx,(500,650))
cv2.imshow('form'+str(form_number),imx)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
